Image manipulation using OpenCV ,Converting a Image into Sketch

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Dog.jpeg")
image
```

```python
plt.imshow(image[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

```python
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.axis(False)
plt.show()
```

```python
grey_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image)
plt.axis(False)
plt.show()
```

```python
invert_image=cv2.bitwise_not(grey_image)
plt.imshow(invert_image)
plt.axis(False)
plt.show()
```

```python
blur_image=cv2.GaussianBlur(invert_image, (111,111),0)
plt.imshow(blur_image)
plt.axis(False)
plt.show()
```

```python
invblur_image=cv2.bitwise_not(blur_image)
plt.imshow(invblur_image)
plt.axis(False)
plt.show()

```

```python
sketch_image=cv2.divide(grey_image,invblur_image, scale=256.0)
plt.imshow(sketch_image)
plt.axis(False)
plt.show()
```

```python
cv2.imwrite('Dog.jpeg', sketch_image)
```

```python
rgb_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_image)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
